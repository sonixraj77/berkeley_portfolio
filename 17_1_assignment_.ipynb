```python
# ===========================
# 1. INSTALL AND IMPORT
# ===========================
# In your terminal or notebook cell, install ucimlrepo:
# !pip install ucimlrepo

from ucimlrepo import fetch_ucirepo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.metrics import confusion_matrix, classification_report

import warnings
warnings.filterwarnings('ignore')
sns.set(style="whitegrid", font_scale=1.1)

# ===========================
# 2. FETCH DATASET
# ===========================
bank_marketing = fetch_ucirepo(id=222)  # This fetches the 'Bank Marketing' dataset from UCI

# Features (X) and Targets (y) as pandas DataFrames
X = bank_marketing.data.features
y = bank_marketing.data.targets

# Optional: Print metadata to see dataset details
print("=== Dataset Metadata ===")
print(bank_marketing.metadata)

print("\n=== Variable Information ===")
print(bank_marketing.variables)

# Preview the features and target
print("\nFeatures (X) shape:", X.shape)
print("Target (y) shape:", y.shape)

# Combine into a single DataFrame for easier manipulation (optional)
df = pd.concat([X, y], axis=1)
df.columns = list(X.columns) + ["y"]  # rename last column to 'y' if not already
print("\nSample of combined DataFrame:")
display(df.head())

# ===========================
# 3. DATA EXPLORATION
# ===========================
print("\n--- Basic Data Info ---")
df.info()

print("\n--- Descriptive Stats (Numeric Columns) ---")
display(df.describe())

print("\n--- Class Distribution for Target 'y' ---")
display(df['y'].value_counts())

# Example: quick visualization of the target distribution
plt.figure(figsize=(4,4))
df['y'].value_counts().plot(kind='bar', color=['lightgreen', 'salmon'], edgecolor='black')
plt.title("Target Variable Distribution (y)")
plt.xlabel("Subscription (no/yes)")
plt.ylabel("Count")
plt.show()

# ===========================
# 4. DATA PREPARATION
# ===========================
# (Below is an example of how you might prepare the data.
#  Adjust or remove steps that arenâ€™t relevant to your approach.)

# Convert target from 'yes/no' to binary (1/0) if needed
if df['y'].dtype == 'object':
    df['y'] = df['y'].map({'yes': 1, 'no': 0})

# Identify categorical vs numeric features
cat_cols = df.select_dtypes(include=['object']).columns
num_cols = df.select_dtypes(include=[np.number]).columns.drop('y', errors='ignore')

# One-hot encode the categorical columns
df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)

# Split features & target
X_encoded = df_encoded.drop('y', axis=1)
y_encoded = df_encoded['y']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)
print("\nShapes:")
print("X_train:", X_train.shape, "X_test:", X_test.shape)
print("y_train:", y_train.shape, "y_test:", y_test.shape)

# Scale numeric columns if needed (especially for KNN or SVM)
scaler = StandardScaler()
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

# Identify numeric columns after encoding (some categorical columns might have become numeric dummies)
# We'll scale only the originally numeric columns or whichever columns you want to standardize
columns_to_scale = [col for col in X_train_scaled.columns if col in num_cols]
X_train_scaled[columns_to_scale] = scaler.fit_transform(X_train_scaled[columns_to_scale])
X_test_scaled[columns_to_scale] = scaler.transform(X_test_scaled[columns_to_scale])

# ===========================
# 5. MODEL TRAINING & EVAL
# ===========================

def evaluate_model(model, X_test, y_test, name="Model"):
    """Utility to print out multiple metrics for easy comparison."""
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # ROC-AUC score requires probability estimates (if supported) or decision function
    if hasattr(model, "predict_proba"):
        y_probs = model.predict_proba(X_test)[:,1]
        auc = roc_auc_score(y_test, y_probs)
    else:
        if hasattr(model, "decision_function"):
            y_decision = model.decision_function(X_test)
            auc = roc_auc_score(y_test, y_decision)
        else:
            auc = float('nan')

    print(f"\n--- {name} Evaluation ---")
    print(f"Accuracy : {acc:.3f}")
    print(f"Precision: {prec:.3f}")
    print(f"Recall   : {rec:.3f}")
    print(f"F1-Score : {f1:.3f}")
    print(f"ROC-AUC  : {auc:.3f}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

# 5.1 KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
evaluate_model(knn, X_test_scaled, y_test, "KNN")

# 5.2 Logistic Regression
logreg = LogisticRegression(solver='liblinear', random_state=42)
logreg.fit(X_train_scaled, y_train)
evaluate_model(logreg, X_test_scaled, y_test, "Logistic Regression")

# 5.3 Decision Tree
dt = DecisionTreeClassifier(random_state=42, max_depth=5)
dt.fit(X_train, y_train)  # not scaled, decision trees handle raw data well
evaluate_model(dt, X_test, y_test, "Decision Tree")

# 5.4 SVM
svm_clf = SVC(kernel='rbf', probability=True, random_state=42)
svm_clf.fit(X_train_scaled, y_train)
evaluate_model(svm_clf, X_test_scaled, y_test, "SVM (RBF)")

# ===========================
# 6. MODEL COMPARISON
# ===========================
models = {
    "KNN": knn,
    "Logistic Regression": logreg,
    "Decision Tree": dt,
    "SVM (RBF)": svm_clf
}

results = []
for model_name, model in models.items():
    if model_name in ["KNN", "Logistic Regression", "SVM (RBF)"]:
        y_pred = model.predict(X_test_scaled)
        if hasattr(model, "predict_proba"):
            y_probs = model.predict_proba(X_test_scaled)[:, 1]
        else:
            # For SVM with no predict_proba, fallback
            y_probs = model.decision_function(X_test_scaled)
    else:  # Decision Tree not scaled
        y_pred = model.predict(X_test)
        y_probs = model.predict_proba(X_test)[:, 1]

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_probs)

    results.append((model_name, acc, prec, rec, f1, auc))

results_df = pd.DataFrame(results, columns=["Model","Accuracy","Precision","Recall","F1-Score","ROC-AUC"])
print("\nModel Comparison:")
display(results_df.sort_values(by="Accuracy", ascending=False))

# ===========================
# 7. INSIGHTS & NEXT STEPS
# ===========================
"""
1. Choose model based on business priorities:
   - High recall if the bank wants to contact as many potential subscribers as possible,
   - High precision if the bank wants to minimize contacting uninterested clients.

2. Possible improvements:
   - Hyperparameter tuning (GridSearchCV, RandomizedSearchCV),
   - SMOTE or class-weight balancing if 'yes' is too rare,
   - Further feature engineering or domain-specific transformations.
3. Consider ensemble methods (Random Forest, XGBoost, etc.) for potentially higher accuracy.

"""
print("\nAll done! Review the above results to choose the best model.")
```
